# Importing Data

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Movie_regression.csv", header = 0)
```

we have 494 values in 'Time_taken' variable, so we'll have to fill the missing values

we can fill the missing values with the mean of the 'Time_taken' variable just for the ease of calculation

# Treating missing values

```python
df['Time_taken'].mean()  #first we calculate the mean
```

```python
df['Time_taken'].fillna(value = df['Time_taken'].mean(), inplace = True) #we use fillna to fill missing values, value will take mean and inplace will permanently replace the blank values with our mean values in the dataframe
```

```python
df.info()  #now we check and see that the 'Time_taken' variable has 506 non-null values and the missing values were filled
```

# Creating dummy variable

Now we want to create some dummy variables to convert our categorical data into numerical data

```python
df.head()
```

We have 2 categorical data under: '3D_available' and 'Genre'

So, '3D_available' has Yes and No whereas 'Genre' has Thriller, Drama, Comedy, Action and we'll always take n-1 to create dummy variables

```python
df = pd.get_dummies(df, columns = ["3D_available", "Genre"], drop_first = True, dtype = int)
```

```python
df.head()
```

# Splitting data into X and Y variables

```python
X = df.loc[:, df.columns != "Collection"]  #choosing all rows with : and all columns except Collection column
```

```python
X.shape
```

```python
X.head()
```

```python
y = df["Collection"] #only choosing collection column
```

```python
y.head()
```

```python
y.shape
```

# Train and test data split

```python
#only 20% of data is used as test data, and we'll train our model with 80% of data as our train data
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
X_train.head() #just checking our X_train data, here we can notice that our index is changed against each row 
```

```python
X_train.shape # X_train will be used to train our model
```

```python
X_test.shape # X_test will be used to check the performance of our model
```

# Training regression tree

```python
from sklearn import tree
```

```python
regtree = tree.DecisionTreeRegressor(max_depth = 3) #never choose max depth greater than 5
```

```python
regtree.fit (X_train, y_train)   #fitting the model
```

# Predict the values using trained model

```python
y_train_pred = regtree.predict(X_train)
```

```python
y_test_pred = regtree.predict(X_test)
```

```python
y_test_pred
```

# Evaluating model performance

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mean_squared_error(y_test, y_test_pred)
```

```python
r2_score(y_train, y_train_pred)
```

```python
r2_score(y_test, y_test_pred)  #use test data to evaluate model performance
```

```python
# always r2 for train data set > r2 for test data set
```
